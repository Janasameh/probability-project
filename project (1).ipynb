```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew 
dataset = pd.read_csv(r'C:\Users\lenovo\Desktop\Diabetes Dataset.csv')
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
for column in columns:
    plt.figure(figsize=(8, 6))
    plt.hist(dataset[column], bins=20, color='blue', edgecolor='black')
    skewness = skew(dataset[column])
    plt.title(f'{column} Distribution (Skewness: {skewness:.2f})')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
df = pd.read_csv(r'C:\Users\lenovo\Desktop\Diabetes Dataset.csv')
dataset = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Outcome', 'Age','DiabetesPedigreeFunction']
subset_df = df[dataset]
correlation_matrix = subset_df.corr()
plt.figure(figsize=(10, 8))
heatmap = plt.pcolor(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(heatmap)
plt.xticks(np.arange(0.5, len(correlation_matrix.columns), 1), correlation_matrix.columns, rotation=45)
plt.yticks(np.arange(0.5, len(correlation_matrix.index), 1), correlation_matrix.index)
plt.title('Heatmap ')
plt.show()
```

```python
  import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statistics
dataset = pd.read_csv(r'C:\Users\lenovo\Desktop\Diabetes Dataset.csv')
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age','Outcome']
for column in columns:
    column_data = dataset[column_name]
    std_dev = np.std(column_data)
    variance = np.var(column_data)
    print(f"Column: {column_name}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Variance: {variance:.2f}")
    print("-" * 30)
plt.figure(figsize=(12, 6))
dataset.boxplot()
plt.title('Boxplot of Columns')
plt.ylabel('Values')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statistics
dataset = pd.read_csv(r'C:\Users\lenovo\Desktop\Diabetes Dataset.csv')
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age','Outcome']
for column_name in dataset.columns:
    column_data = dataset[column_name]
    min_value = np.min(column_data)
    max_value = np.max(column_data)
    print(f"Column: {column_name}")
    print(f"Minimum: {min_value:.2f}")
    print(f"Maximum: {max_value:.2f}")
    print("-" * 30)
plt.figure(figsize=(12, 6))
dataset.boxplot()
plt.title('Boxplot of Columns')
plt.ylabel('Values')
plt.show()
```
